```python
def ricodifica(x):
    if x == "":
        return ""
    else:
        return str(x)
        
```

```python
codici_regionali = pd.read_excel('Glossario_SIOPE_Region2013.xls', "SPESE", skiprows = 5)
```

```python
codici_regionali_1 = codici_regionali 
codici_regionali_1.fillna("", inplace = True)
```

```python
codici_regionali_1["CODICI GESTIONALI_5digit"] = codici_regionali_1["CODICE BILANCIO"].apply(lambda x: ricodifica(x)) \
 + codici_regionali_1["Unnamed: 1"].apply(lambda x: ricodifica(x)) + codici_regionali_1["Unnamed: 2"].apply(lambda x: ricodifica(x)) 
    
```

```python
codici_regionali_1_ridotto =  codici_regionali_1.iloc[1:,:]
```

```python
codici_regionali_1_ridotto2 = codici_regionali_1_ridotto[codici_regionali_1_ridotto["CODICI GESTIONALI_5digit"] != ""]
```

```python
#codici_regionali_1_ridotto2 
```

```python
codici_regionali_1_ridotto2.columns = [u'CODICE BILANCIO I digit', u'CODICE BILANCIO II-III digit', u'CODICE BILANCIO IV-V digit', 
                                       u'CODICI GESTIONALI', u'DESCRIZIONI VOCI', u'DESCRIZIONI VOCI 5 digit',
                                       u'DEFINIZIONI 5 digit', u'CODICI raccordo_5digit']
```

```python
codici_bilancio4 = codici_regionali_1_ridotto2[codici_regionali_1_ridotto2["CODICI GESTIONALI"] != ""]

codici_bilancio4["CODICI raccordo livello 1"] = codici_bilancio4["CODICI raccordo_5digit"].map(lambda x: x[0])
codici_bilancio4["CODICI raccordo livello 2"] = codici_bilancio4["CODICI raccordo_5digit"].map(lambda x: x[0:3])
codici_bilancio4["CODICI raccordo livello 3"] = codici_bilancio4["CODICI raccordo_5digit"].map(lambda x: x)
```

```python
#codici_bilancio4
```

```python
codici_bilancioI_II_III = codici_regionali_1_ridotto2[codici_regionali_1_ridotto2["CODICI GESTIONALI"] == ""]
```

```python
codici_bilancioI_II_III_ridotto = codici_bilancioI_II_III[["CODICI raccordo_5digit", "DESCRIZIONI VOCI", "DEFINIZIONI 5 digit"]]
```

```python
codici_bilancioI_II_III_ridotto.columns = [u'CODICI', u'DESCRIZIONI VOCI', u'DEFINIZIONI']
```

```python
#codici_bilancioI_II_III_ridotto
```

```python
PassaggioI = pd.merge(codici_bilancio4, codici_bilancioI_II_III_ridotto, left_on = "CODICI raccordo livello 1", right_on = "CODICI", suffixes = ("", " livello1"), how="left")
PassaggioII = pd.merge(PassaggioI, codici_bilancioI_II_III_ridotto, left_on = "CODICI raccordo livello 2", right_on = "CODICI", suffixes = ("", " livello2"), how="left")
PassaggioIII = pd.merge(PassaggioII, codici_bilancioI_II_III_ridotto, left_on = "CODICI raccordo livello 3", right_on = "CODICI", suffixes = ("", " livello3"), how="left")
```

```python
raccordo_codici_regionale = PassaggioIII
```

```python
raccordo_codici_regionale["CODICI GESTIONALI"] = raccordo_codici_regionale["CODICI GESTIONALI"].map(lambda x: unicode(x)[:-2])
```

```python
raccordo_codici_regionale["DESCRIZIONI VOCI 5 digit"] = raccordo_codici_regionale["DESCRIZIONI VOCI"] + raccordo_codici_regionale["DESCRIZIONI VOCI 5 digit"]
```

```python
raccordo_codici_regionale.loc[255] = ""
raccordo_codici_regionale.iloc[255, 3] = "9997"
raccordo_codici_regionale.iloc[255, 5:] = "TITOLO 5: altro" 

raccordo_codici_regionale.loc[256] = ""
raccordo_codici_regionale.iloc[256, 3] = "9998"
raccordo_codici_regionale.iloc[256, 5:] = "TITOLO 5: altro" 
raccordo_codici_regionale.loc[257] = ""
raccordo_codici_regionale.iloc[257, 3] = "9999"
raccordo_codici_regionale.iloc[257, 5:] = "TITOLO 5: altro" 
```

```python
raccordo_codici_regionale["DESCRIZIONI VOCI livello2"].fillna("9999", inplace = True)
raccordo_codici_regionale["DESCRIZIONI VOCI livello3"].fillna("9999", inplace = True)

#raccordo_codici_regionale.to_excel("raccordo_codici_regionale.xlsx")
raccordo_codici_regionale
```

```python
con = sqlite3.connect(db)
cursor = con.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```
